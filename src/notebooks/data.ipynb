## EDA: Word count for ViLexNorm splits

```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style='whitegrid')

base_dir = Path('.')
files = {
    'train': base_dir / 'vilexnorm' / 'ViLexNorm_train.json',
    'dev': base_dir / 'vilexnorm' / 'ViLexNorm_dev.json',
    'test': base_dir / 'vilexnorm' / 'ViLexNorm_test.json',
}

def word_count(text):
    return len(str(text).split())

records = []
for split_name, file_path in files.items():
    rows = load_json(str(file_path))
    for row in rows:
        text = row.get('normalized', row.get('original', ''))
        records.append({
            'split': split_name,
            'word_count': word_count(text),
        })

df = pd.DataFrame(records)
summary = (
    df.groupby('split')['word_count']
    .agg(['count', 'mean', 'median', 'min', 'max'])
    .round(2)
    .sort_index()
)

print('Summary statistics for word counts based on the normalized text:')
print(summary)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

sns.boxplot(data=df, x='split', y='word_count', hue='split', dodge=False, ax=axes[0], palette='Set2', legend=False)
axes[0].set_title('Word count distribution by split')
axes[0].set_xlabel('Split')
axes[0].set_ylabel('Words per sample')

sns.histplot(
    data=df,
    x='word_count',
    hue='split',
    bins=40,
    element='step',
    stat='density',
    common_norm=False,
    ax=axes[1],
)
axes[1].set_title('Word count histogram by split')
axes[1].set_xlabel('Words per sample')
axes[1].set_ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
def get_comments_from_urls(urls, client_id='client_51gm9xxri_100009489619122'):
    import requests

    api_url = 'https://api.fbaio.xyz/call'
    headers = {'Content-Type': 'application/json'}
    all_comments = []

    for post_url in urls:
        payload = {
            'id': client_id,
            'apiname': 'get_list_fb_comment',
            'apiparams': {
                'url': post_url,
                'type': 'CHRONOLOGICAL_UNFILTERED_INTENT_V1',
                'cursor': ''
            }
        }

        try:
            response = requests.post(api_url, json=payload, headers=headers, timeout=30)
            response.raise_for_status()
            comments = response.json().get('result', {}).get('comments', [])
            all_comments.extend(
                comment.get('text', '')
                for comment in comments
                if comment.get('text')
            )
        except Exception as e:
            print(f'Khong lay duoc comment tu URL: {post_url} | Loi: {e}')

    return all_comments

# Vi du:
# comments = get_comments_from_urls(post_urls[:3])
# len(comments), comments[:5]
```
